```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
```

```python
basic="https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_HistoryAutoSuggest_1_1_na_na_na&otracker1=AS_Query_HistoryAutoSuggest_1_1_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobiles&requestId=80368ab3-5cbe-46cd-8444-45b4e1e9f3e8&as-backfill=on&sort=price_desc&page="
```

```python
ram_info=[]
info=[]
price=[]
mrp=[]
rating=[]
for i in range(1,38):
    pageno=str(i)
    url=basic+pageno
    website=requests.get(url)
    soup=BeautifulSoup(website.text)
    ram=soup.find_all('div',class_="fMghEO")
    scrap=soup.find_all('div',class_="_4rR01T")
    scrap_price=soup.find_all('div',class_="_30jeq3 _1_WHN1")
   
    for i in ram:
        ram_info.append(i.text)
        
    for i in scrap:
        info.append(i.text)
        
    for i in scrap_price:
        r=i.text
        if r:
            price.append(r)
        else:
            price.append(nan)
            
        
```

```python
brand=[]
model=[]
color=[]
rom=[]
ram=[]
display=[]
frontcam=[]
battery=[]
warranty=[]
rearcam=[]

for i in info:
    brand.append(i.split(' ')[0])
    

for j in info:
    model.append(j.split(' ')[1])
    

for p in info:
    z=re.findall(r'\((\w+\s\w+|\w*)',p)
    if z:
        color.append(z)
    else:
        color.append(np.nan)
        
        
for k in ram_info:
        a=re.findall(r"\d*\sGB\sRAM",k)
        if a:
            ram.append(a) 
        else:
            ram.append(np.nan)
            
            
for m in ram_info:
    rom.append(re.findall(r'(\d*\sGB)\sROM',m))
    

for n in ram_info:
    z=re.findall(r'(\d.\d*)\sinch',n)
    if z:
        display.append(z)
    else:
        display.append(np.nan)
        
    
for o in ram_info:
    q=re.findall(r'(\d*MP)\sFront\sCamera',o)
    if q:
        frontcam.append(q)
    else:
        frontcam.append(np.NaN)
        
        
for i in ram_info:
    bat=re.findall(r'(\d*)\smAh',i)
    if bat:
        battery.append(bat)
    else:
        battery.append(np.nan)
        
    
for k in ram_info:
    sim=re.findall(r'(\d)\sYear',k)
    if sim:
        warranty.append(sim)
    else:
        warranty.append(np.nan)
        
        
for r in ram_info:
    pik=re.findall(r'play(\d*)MP',r)
    if pik:
        rearcam.append(pik)
    else:
        rearcam.append(np.nan)
        
```

```python
print(len(model))
print(len(brand))
print(len(rom))
print(len(ram))
print(len(color))
print(len(price))
print(len(display))
print(len(frontcam))
print(len(rearcam))
print(len(battery))
print(len(warranty))
```

```python
df=pd.DataFrame({'Brand':brand,
                 'Model Name':model,
                 'Colour':color,
                 'Price':price,
                 'RAM':ram,
                 'ROM':rom,
                 'Display':display,
                 'Front Camera':frontcam,
                 'Rear Camera':rearcam,
                 'Battery':battery,
                 'Warranty Period':warranty                
})
```
